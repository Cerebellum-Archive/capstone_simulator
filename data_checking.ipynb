```python
spdr = ['SPY','XLK', 'XLF', 'XLV', 'XLY', 'XLP', 'XLE', 'XLI', 'XLB', 'XLRE', 'XLU']
```

```python
def log_returns(df):
    df = np.log(df)
    df = df - df.shift(1)
    return df
```

```python
## Old Dataset pulled in 2018 from yf

spyders = pd.read_csv('SPYDERS.csv')
spyders.Date = pd.DatetimeIndex(spyders['Date'])
spyders.set_index(spyders.Date,inplace=True)
X_level = spyders.drop('Date',axis=1)
X_log_returns = log_returns(X_level).shift(1)
X_log_returns
```

```python
#  New datasert pulled in 11/26/2023 from yf

X_level_new = yf.download(spdr, start="2000-01-03", end="2016-08-06")['Adj Close']
X_log_returns_new = log_returns(X_level_new).shift(1)
X_log_returns_new
```

```python
import yfinance as yf

def get_stock_data(ticker, start_date, end_date):
    # Download historical data
    data = yf.download(ticker, start=start_date, end=end_date)

    # Extract adjusted closing prices
    adjusted_prices = data

    # Get dividend data
    dividends = yf.Ticker(ticker).dividends

    return adjusted_prices, dividends

if __name__ == "__main__":
    # Set the stock ticker symbol, start date, and end date
    ticker_symbol = "XLF"  # Example: Apple Inc.
    start_date = "1990-01-01"
    end_date = "2023-01-01"

    # Get adjusted closing prices and dividends
    adjusted_prices, dividends = get_stock_data(ticker_symbol, start_date, end_date)

    # Print the results
    print("Adjusted Prices:")
    print(adjusted_prices)

    print("\nDividends:")
    print(dividends)
```
